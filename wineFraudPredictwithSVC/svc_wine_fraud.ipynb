# Project Title: Using Support Vector Machine Classification Algorithm, Predict if a Wine is Fraud or Legit based on Chemical Composition

## Project Background

Frauds in the wine industry are very real?
You may be wondering if I can share any proof of that. Here are some links for you to read more about it:

- https://www.theguardian.com/global/2016/sep/11/the-great-wine-fraud-a-vintage-swindle
- https://daily.sevenfifty.com/beyond-high-profile-scandals-wine-fraud-is-rampant/

I think you got the idea. The next natural question may be what type of frauds in the wine industry are there?

As per Wikipedia (https://en.wikipedia.org/wiki/Wine_fraud), the most common one is where wines are adulterated with inferior ones and sold as premium ones. Relabeling is another one and there are scams related to wine investments.

The main goal of this supervised machine learning project is to use **Support Vector Machine Classification algorithm** to predict if a wine is 'legit' or a 'fraudulent' one based on it's chemical composition.

## Project Methodology

## Problem Identification

**Hypothesis:** Predict a wine if it is a 'legit' or a 'fraud' type based on the chemical composition utilizing Support Vector Machine (SVM) Classification, a supervised machine learning algorithm

**Context for Wine Chemical Composition Usage:** Quality can be a subjective choice, so to ensure a consistent adoption of wine quality there are some important chemical composition requirements which decides the wine quality and some of these characteristics are acidity, residual sugar, density, ph and much more.

**Data Source:** In order to support machine learning algorithm, UCI Machine Learning Respository has made the data available for specific to this purpose which will be utilized for this project. This data is available at https://archive.ics.uci.edu/ml/datasets/wine+quality

# Pre-Processing and Training Data Development

```python
# Importing required packages and libraries
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# Using pandas dummy variables first for the type column
df['type'] = pd.get_dummies(df['type'],drop_first=True)
```

```python
df.tail()
```

Here 'red' wine is categorized as '0' and 'white' wine is categorized as '1'.

```python
# Defining X and y
X = df.drop('quality',axis=1)
y = df['quality']
```

## Stratify Target Variable 'y' due to Data Imbalancing

```python
# Splitting into 80% training and 30% testing dataset
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=101, stratify=y)
```

## Scaling Dataset for Machine Learning

```python
# Using standard scaler for scaling features
scaler = StandardScaler()
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

```python
# Shape of scaled_X_test
scaled_X_test.shape
```

Now dataset is ready for the machine learning modeling purpose.

## Execution Plan for SVC Modeling for the Imbalanced Data

As number of frauds are normally less in the overall dataset, data becomes imbalanced. To get better results, I am going to experiment with different techniques to see where do I get the best results.

Here is the summary of the steps I plan to execute:

- Stratify for the target variable ('y') - Already completed
- Base or Default Support Vector Machine (SVM) Binary Classification model with default hyper parameters
- SVM Classifier model with **class_weight='balanced'** due to imbalanced data
- GridSearchCV for Best Hyper Parameters for SVM Classifier
- SVM Classifier model with **class_weight='balanced' and Undersampling Data**
- SVM Classifier model with **class_weight='balanced' and Oversampling Data**
- SVM Classifier model with **class_weight='balanced' and SMOTE Data**

# Machine Learning Modeling

```python
# Importing required packages and libraries
from sklearn.svm import SVC 
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import GridSearchCV
```

## Base Support Vector Machine (SVM) Classifier with Default Hyper Parameters

```python
# Modeling for the default base scenario
svc_default = SVC()
# Fitting scaled training data set
svc_default.fit(scaled_X_train,y_train)

# predicting for the default model
pred_default = svc_default.predict(scaled_X_test)
```

```python
# Printing confusion matrix for default base scenario
confusion_matrix(y_test,pred_default)
```

```python
# Printing classification report for default base scenario
print("Classification Report for Base Modeling:\n", classification_report(y_test, pred_default))
```

This is not a good scenario as we can't identify any frauds here.

## SVM Classifier with Class Weight ("balanced") and other default Hyper Parameters

For **Class Imbalanced" such as this scenario of number of frauds-related data relatively much less than the normal data (also known as "imbalanced data"), Support Vector Machine has a mechanism to use class weight parameter as "balanced" where data is imbalanced.** 

"When the class_weights = ‘balanced’, the model automatically assigns the class weights inversely proportional to their respective frequencies."

```python
# Modeling for the base scenario with class weights
svc_weight = SVC(class_weight='balanced')
# Fitting scaled training data set
svc_weight.fit(scaled_X_train,y_train)

# predicting for the default model
pred_weight = svc_weight.predict(scaled_X_test)
```

```python
# Printing confusion matrix for base scenario with class weights
confusion_matrix(y_test,pred_weight)
```

```python
# Printing classification report for base scenario with class weights
print("Classification Report for Class Weight Modeling:\n", classification_report(y_test, pred_weight))
```

This is much better than the default scenario for the fraud idenfications.

## Using Data Balancing Techniques

Due to imbalanced data, it is important to check which data balancing technique will work better for this dataset:

- Upsampling or Over-sampling
- Downsampling or Under-sampling
- Synthetic Minority Oversampling Technique (SMOTE) - a type of oversampling where synethic data is generated for the minority label to compensate for the higher presence of majority label

I will try the above 3 different scenario to see if it helps to predict better results with no stratification for y.

```python
# Importing required packages and libraries
from imblearn.over_sampling import SMOTE
from sklearn.utils import resample
```

```python
# Creating testing and training dataset again with no stratify parameter and test size of 30%
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
# concating training dataset
df_train = pd.concat([X_train, y_train],axis=1)
```

```python
# Finding the Legits and Frauds in the training data set
legits = df_train[df_train['quality']=='Legit']
frauds = df_train[df_train['quality']=='Fraud']

print("Number of legitimate wines before balancing in training set:", len(legits.value_counts()))
print("Number of fraudulent wines before balancing in training set:", len(frauds.value_counts()))
```

### SVM Classifier Modeling after Balacing with Upsampling or Oversampling

```python
# Upsampling the dataset
upsampled_frauds = resample(frauds, replace=True, n_samples=len(legits), random_state=21)
```

```python
# Combining both legits and unsampled_frauds to check if we have the good distribution of frauds
upsampled_quality = pd.concat([legits, upsampled_frauds])
```

```python
# Checking for the frauds
upsampled_quality['quality'].value_counts()
```

```python
# Separating into X and y from up sampled data
X_train_up = upsampled_quality.drop('quality', axis=1)
y_train_up = upsampled_quality['quality']
```

```python
# Scaling X_train_up
scaler = StandardScaler()
X_train_upscaled = scaler.fit_transform(X_train_up)
X_test_upscaled = scaler.transform(X_test)
```

```python
# SVC with upsampled data
svc_upsampled = SVC()
svc_upsampled.fit(X_train_upscaled,y_train_up)
```

```python
# Predicting with upsampled model
pred_upsampled = svc_upsampled.predict(X_test_upscaled)
```

```python
# Printing confusion matrix for the upscaled data
confusion_matrix(y_test,pred_upsampled)
```

```python
# Printing classification report for upsampled data
print("Classification Report for upsampled data:\n", classification_report(y_test, pred_upsampled))
```

### SVM Classifier Modeling after Balacing with Downsampling or Undersampling

```python
# Downsampling the dataset
downsampled_legits = resample(legits, replace=False, n_samples=len(frauds), random_state=21)
```

```python
# concat downsampled legits with frauds
downsampled_quality = pd.concat([downsampled_legits,frauds])
```

```python
# Checking for the frauds
downsampled_quality['quality'].value_counts()
```

```python
# Separating into X and y from down sampled data
X_train_down = downsampled_quality.drop('quality', axis=1)
y_train_down = downsampled_quality['quality']
```

```python
# Scaling X_train_down
scaler = StandardScaler()
X_train_downscaled = scaler.fit_transform(X_train_down)
X_test_downscaled = scaler.transform(X_test)
```

```python
# SVC with downsampled data
svc_downsampled = SVC()
svc_downsampled.fit(X_train_down,y_train_down)
```

```python
# Predicting with downsampled model
pred_downsampled = svc_downsampled.predict(X_test_downscaled)
```

```python
# Printing confusion matrix for the downscaled data
confusion_matrix(y_test,pred_downsampled)
```

```python
# Printing classification report for downsampled data
print("Classification Report for downsampled data:\n", classification_report(y_test, pred_downsampled))
```

### SVM Classifier Modeling after Balacing with SMOTE

```python
# Generating synthetic samples
sm = SMOTE(random_state=21)
X_smtrain, y_smtrain = sm.fit_resample(X_train, y_train)
```

```python
# Scaling X_smtrain
scaler = StandardScaler()
X_train_smote = scaler.fit_transform(X_smtrain)
X_test_smote = scaler.transform(X_test)
```

```python
# SVC with SMOTE data
svc_smote = SVC()
svc_smote.fit(X_train_smote,y_smtrain)
```

```python
# Predicting with SMOTE model
pred_smote = svc_smote.predict(X_test_smote)
```

```python
# Printing confusion matrix for the SMOTE data
confusion_matrix(y_test,pred_smote)
```

```python
# Printing classification report for SMOTE data
print("Classification Report for SMOTE data:\n", classification_report(y_test, pred_smote))
```

The best results we have received so far are related to running the SVM with class weights as 'balanced'. After analyzing so many scenario, the main final scenario is what if we run the class weights for the original dataset with different class weights values to see if we can get the better results.

## SVM Classifier with Different Class Weights Hyper Parameters Using Grid Search

```python
# Modeling for the different class weights from 0 to 1 and Setting the range for class weights
weights = np.linspace(0.0,0.99,200)

#Creating a dictionary grid for grid search
param_grid = {'class_weight': [{'Legit':x, 'Fraud':1.0-x} for x in weights]}

svc = SVC()
# Creating grid search
grid_manual = GridSearchCV(svc, param_grid)
```
