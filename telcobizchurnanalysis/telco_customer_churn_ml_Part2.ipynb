# Project Title: Predict Customer Churn for a Telco Business based on Customer Segmentation (Churn & Cohort Analysis) using different Machine Learning algorithms - Part 2 of 3

# Project Background and Problem Identification

As per Wikipedia (https://en.wikipedia.org/wiki/Churn_rate), the customer churn rate means number of subscribers leaving a service provider or supplier during a given time period. This is also known as customer attrition or customer turnover or customer defection (https://en.wikipedia.org/wiki/Customer_attrition). In a simpler terms, this is simply loss of customers to a business. This is an industry-wide problem more prominent in certain segments such as telecom, entertainment industry etc but most of the service providers are dealing with this problem all the time.

In order to find a proactive way, many scientists and experts are working on various machine learning models to keep fine tuning their models so that they can proactively predict the customer attrition. If the service provider can get the signal ahead of time, they can put up the necessary plans to reduce the churn rate.

**The goal of this project is to try different machine learning algorithms and see which one fits the best.**

This project has been divided into 3 Jupyter Notebooks as per following:

- First Jupyter Notebook: Full step-by-step EDA, Pre-Processing and Maachine Learning modeling covering Logistic Regression, Decision Tree Classifier, Random Forest, AdaBoost, Gradient Boosting, KNN, and Support Vector Classifier without any type of data balancing and with default hyperparameters

- **Second Jupyter Notebook: Focused on data balancing i.e. oversampling, undersampling and SMOTE and then running those algorithms which performed better in the first Jupyter Notebook and these are Logistic Regression, AdaBoost, Gradient Boosting, and Support Vector Classifier with default hyperparameters**

- Third Jupyter Notebook: Focused on the finally selected machine learning algorithms and improving the models with hyperparameter tuning

**This is the Second Jupyter Notebook for this project.**

## Creation of Broader Cohort Groups based on Tenure

I am going to create a new column for the following 4 categories based on tenure grouping:

- '0-12Months'
- '12-24Months'
- '24-48Months'
- 'Over48Months'

```python
# Checking tenure series
df['tenure']
```

```python
# creating grouping based on tenure by defining a function
def cohort(tenure):
    if tenure<13:
        return '0-12Months'
    elif tenure<25:
        return '12-24Months'
    elif tenure<49:
        return '24-48Months'
    else:
        return 'Over48Months'
```

```python
# Applying the above function to tenure
df['tenure_cohort'] = df['tenure'].apply(cohort)
df[['tenure_cohort', 'tenure']]
```

```python
# to further analyze, create a countplot for tenure_cohort
sns.countplot(x='tenure_cohort',data=df, hue='Churn');
```

# Pre-Processing and Training Data Development

To work on the tree-based algorithms, I don't need to scale the data but I plan to test out not only tree-based models but also distance-based models. Hence I am going to scale the data. Also I need to convert to numeric type for the machine learning purpose after dropping the customerID column as that is not adding any value to the modeling.

```python
# Importing required packages
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# Splitting into X and y first
X = df.drop(['customerID','Churn'], axis=1)
y = df['Churn']
```

```python
# Converting categorical to numerical type
X = pd.get_dummies(X, drop_first=True)
```

```python
# Splitting into training and testing (10%) data set
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.1,random_state=101)
```

## Data Balancing

```python
# Importing required packages
from sklearn.utils import resample
from imblearn.over_sampling import SMOTE
```

```python
# Combining X_scaled_train with y_train for the data balancing purpose
df_train = pd.concat([X_train,y_train], axis=1)
```

```python
# Checking df_train
df_train.head()
```

### Balancing with Upsampling or Oversampling

```python
# Balancing with upsampling for churn
no_churn = df_train[df['Churn']=='No']
yes_churn = df_train[df['Churn']=='Yes']

# resample yes_churn
upsampled_yes_churn = resample(yes_churn, replace=True, n_samples=len(no_churn), random_state=21)

# Combining no_churn with upsampled yes_churn
upsampled = pd.concat([no_churn,upsampled_yes_churn])

# counting values for both yes and no churns
upsampled['Churn'].value_counts()
```

```python
# Separating X and y for the training set for upsampled
X_upsampled_train = upsampled.drop('Churn', axis=1)
y_upsampled_train = upsampled['Churn']
```

```python
# Scaling data for the Upsampled
scaler = StandardScaler()
X_up_scaled_train = scaler.fit_transform(X_upsampled_train)
X_scaled_test = scaler.transform(X_test)
```

### Balancing with Downsampling or Undersampling

```python
# resample no_churn
downsampled_no_churn = resample(no_churn, replace=False, n_samples=len(yes_churn), random_state=21)

# Combining yes_churn with downsampled_no_churn
downsampled = pd.concat([downsampled_no_churn,yes_churn])

# counting values for both yes and no churns
downsampled['Churn'].value_counts()
```

```python
# Separating X and y for the training set for downsampled
X_downsampled_train = downsampled.drop('Churn', axis=1)
y_downsampled_train = downsampled['Churn']
```

```python
# Scaling data for the Downsampled
scaler = StandardScaler()
X_down_scaled_train = scaler.fit_transform(X_downsampled_train)
X_scaled_test = scaler.transform(X_test)
```

### Balancing with SMOTE

```python
# Creating SMOTE model
sm = SMOTE(random_state=21)
X_sm_train, y_sm_train = sm.fit_resample(X_train, y_train)

# Combining X_sm_train with y_sm_train
smote_sampled = pd.concat([X_sm_train,y_sm_train], axis=1)

# counting values for both yes and no churns
smote_sampled['Churn'].value_counts()
```

```python
# Scaling data for the SMOTE
scaler = StandardScaler()
X_smote_scaled_train = scaler.fit_transform(X_sm_train)
X_scaled_test = scaler.transform(X_test)
```

# Machine Learning Modeling with Default Hyperparameters for Upsampled, Downsampled and SMOTE Data

```python
# Importing required packages
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import confusion_matrix, classification_report, plot_confusion_matrix, accuracy_score
```

```python
# Creating model results evaluation function
def model_results(model, X_train,y_train,X_test, y_test):
       
    # fitting the model
    model.fit(X_train,y_train)
    # predicting
    preds = model.predict(X_test)
    # results evaluation
    print(classification_report(y_test,preds))
    plot_confusion_matrix(model,X_test,y_test)
```

## Logistic Regression Modeling with default Hyperparameters

```python
# Creating logisitic regression model with default hyperparameters
lr_model = LogisticRegression()
```

### Logistic Regression for Upsampled Data

```python
# Model results for regression model for upsampled data
model_results(lr_model,X_up_scaled_train,y_upsampled_train,X_scaled_test,y_test)
```

### Logistic Regression for Downsampled Data

```python
# Model results for regression model for downsampled data
model_results(lr_model,X_down_scaled_train,y_downsampled_train,X_scaled_test,y_test)
```

### Logistic Regression for SMOTE Data

```python
# Model results for regression model for SMOTE data
model_results(lr_model,X_smote_scaled_train,y_sm_train,X_scaled_test,y_test)
```

**For the above 3 data balancing scenario with logistic regression, SMOTE is the best scenario.**
