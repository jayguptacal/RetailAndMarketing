# Project Title: Predict Customer Churn for a Telco Business based on Customer Segmentation (Churn & Cohort Analysis) using different Machine Learning algorithms - Part 3 of 3

# Project Background and Problem Identification

As per Wikipedia (https://en.wikipedia.org/wiki/Churn_rate), the customer churn rate means number of subscribers leaving a service provider or supplier during a given time period. This is also known as customer attrition or customer turnover or customer defection (https://en.wikipedia.org/wiki/Customer_attrition). In a simpler terms, this is simply loss of customers to a business. This is an industry-wide problem more prominent in certain segments such as telecom, entertainment industry etc but most of the service providers are dealing with this problem all the time.

In order to find a proactive way, many scientists and experts are working on various machine learning models to keep fine tuning their models so that they can proactively predict the customer attrition. If the service provider can get the signal ahead of time, they can put up the necessary plans to reduce the churn rate.

**The goal of this project is to try different machine learning algorithms and see which one fits the best.**

This project has been divided into 3 Jupyter Notebooks as per following:

- First Jupyter Notebook: Full step-by-step EDA, Pre-Processing and Maachine Learning modeling covering Logistic Regression, Decision Tree Classifier, Random Forest, AdaBoost, Gradient Boosting, KNN, and Support Vector Classifier without any type of data balancing and with default hyperparameters

- Second Jupyter Notebook: Focused on data balancing i.e. oversampling, undersampling and SMOTE and then running those algorithms which performed better in the first Jupyter Notebook and these are Logistic Regression, AdaBoost, Gradient Boosting, and Support Vector Classifier with default hyperparameters

- **Third Jupyter Notebook: Focused on the finally selected machine learning algorithms and improving the models with hyperparameter tuning. Based on the performance of various machine learning algorithms in the first and second Jupyter Notebooks, only following will be modeled:

    - **Logistic Regression Classifier**
    - **AdaBoost Classifier**
    - **Gradient Boosting Classifier**
    - **Support Vector Classifier**

**This is the Third Jupyter Notebook for this project.**

## Creation of Broader Cohort Groups based on Tenure

I am going to create a new column for the following 4 categories based on tenure grouping:

- '0-12Months'
- '12-24Months'
- '24-48Months'
- 'Over48Months'

```python
# creating grouping based on tenure by defining a function
def cohort(tenure):
    if tenure<13:
        return '0-12Months'
    elif tenure<25:
        return '12-24Months'
    elif tenure<49:
        return '24-48Months'
    else:
        return 'Over48Months'
```

```python
# Applying the above function to tenure
df['tenure_cohort'] = df['tenure'].apply(cohort)
df[['tenure_cohort', 'tenure']]
```

```python
# to further analyze, create a countplot for tenure_cohort
sns.countplot(x='tenure_cohort',data=df, hue='Churn');
```

# Pre-Processing and Training Data Development

To work on the tree-based algorithms, I don't need to scale the data but I plan to test out not only tree-based models but also distance-based models. Hence I am going to scale the data. Also I need to convert to numeric type for the machine learning purpose after dropping the customerID column as that is not adding any value to the modeling.

```python
# Importing required packages
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# Splitting into X and y first
X = df.drop(['customerID','Churn'], axis=1)
y = df['Churn']
```

```python
# Converting categorical to numerical type
X = pd.get_dummies(X, drop_first=True)
```

```python
# Splitting into training and testing (10%) data set
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.1,random_state=101)
```

```python
# Scaling both X_train and X_test
scaler = StandardScaler()
X_scaled_train = scaler.fit_transform(X_train)
X_scaled_test = scaler.transform(X_test)
```

# Machine Learning Modeling with Hyperparameters Tuning

```python
# Importing required packages
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

from sklearn.metrics import confusion_matrix, classification_report, plot_confusion_matrix, accuracy_score
```

```python
# Creating model results evaluation function
def model_results(model):
       
    # fitting the model
    model.fit(X_train,y_train)
    # predicting
    preds = model.predict(X_test)
    # results evaluation
    print(classification_report(y_test,preds))
    plot_confusion_matrix(model,X_test,y_test)
```

## Logistic Regression Modeling

```python
# Creating logisitic regression model with default hyperparameters
lr_model = LogisticRegression()
```

```python
#help(LogisticRegression())
```

Logistic Regression does not have any critical hyperparameters to tune. So I am going to fune for the following:
- Regularization (penalty) can sometimes be helpful. penalty: [none, l1, l2, elasticnet]
- C: [0.01,0.1,1.0,100]


```python
# Creating grid parameters for logistic classifier
param_grid = {'penalty':['none','l1','l2','elasticnet'],
              'C':[0.01,0.1,1.0,100]}
```

```python
# grid modeling
grid_lr = GridSearchCV(lr_model,param_grid)
grid_lr.fit(X_train,y_train)

preds_grid_lr = grid_lr.predict(X_test)
```

```python
# Finding best parameters
grid_lr.best_params_
```

```python
# creating logistic regression with best hyperparameters
lr_best = LogisticRegression(C=0.01,penalty='l2')
```

```python
# results for the best logistic regression
model_results(lr_best)
```

## AdaBoost Classifier Modeling

```python
#help(AdaBoostClassifier())
```

```python
# Creating AdaBoost classifier model
ab_grid = AdaBoostClassifier()
```

```python
# Creating grid parameters for adaboost classifier
param_grid = {'n_estimators':[10,50,100,500],
             'learning_rate':[0.0001,0.001,0.01,0.1,0.2,1.0]}
```

```python
# grid modeling
grid_ab = GridSearchCV(ab_grid,param_grid)
grid_ab.fit(X_train,y_train)

preds_grid_ab = grid_ab.predict(X_test)
```

```python
# Finding best parameters
grid_ab.best_params_
```

Since above both hyperparameters are the default ones, no further hyperparameter tuning will be done.

```python
# Model resutls for AdaBoost
model_results(ab_grid)
```

## Gradient Boosting Classifier Modeling

```python
# Creating gradient boosting classifier model
gb_model = GradientBoostingClassifier()
```

```python
# Setting up grid parameters
param_grid = {'n_estimators':[50,100,200,500],
             'learning_rate':[0.0001,0.001,0.01,0.05,0.1,1.0],
             'max_depth':[3,4,5]}
```

```python
# grid modeling
grid_gb = GridSearchCV(gb_model,param_grid)
grid_gb.fit(X_train,y_train)

preds_grid_gb = grid_gb.predict(X_test)
```

```python
grid_gb.best_params_
```

```python
# Creating Gradient Boosting model with the best hyperparameters
gb_best = GradientBoostingClassifier(learning_rate=0.05,max_depth=3,n_estimators=100)
```

```python
# Results for the Gradient Boosting mode with the best hyperparameters
model_results(gb_best)
```

## Support Vector Machine Classifier (SVC) Modeling

```python
# Creating SVC model
svc_model = SVC()
```

```python
# Creating parameters for hypertuning
param_grid = {'C':[0.001,0.01,0.1, 0.5, 1.0],
             'gamma':['scale','auto']}
```

```python
# modeling for the grid model for the SVC
grid_svc = GridSearchCV(svc_model, param_grid)
grid_svc.fit(X_train,y_train)

preds_grid_svc = grid_svc.predict(X_test)
```

```python
grid_svc.best_params_
```

```python
# Creating SVC model for the best hyperparameters
svc_best = SVC(C=1.0,gamma='auto')
```

```python
# Results for the SVC with the best hpyerparameters
model_results(svc_best)
```
