While researching for the M-code, I found a good tip where I create a rolling calendar which will be good for the new data based on the current point in time. Here are the steps I followed:

**Create a Rolling Calendar**

    - create a new blank query
    - In the formula bar, generate a starting date by entering a literal (YYYY, MM, DD) such as #date(YYYY,MM,DD) format or =date(2021,08,19)
    - Now click on the fx icon to add a custom step and enter the following formula:
    = List.Dates (Source, Number.From(DateTime.LocalNow()) - Number.From(Source), #duration(1,0,0,0))
    
    - Now convert the resulting list into a table (List Tools - To Table) and format the column as a Date
    - Last step is to add Calculated Date Columns as necessary using Add Column tools

## Data Modeling and Creating Table Relationships

**"Data modeling is one of the features of power BI to connect various data sources based on a relationship."**

In my example case, ProductKey is a common field which can help connect various tables based on this common column feature. This common field is also known as 'Key' for establishing relationship among tables.

**Database Normalization** - While importing various tables and fields, I constantly kept on eye on database normalization as it is important to avoid data duplication to preserve data integrity. This is an important concept to achieve the following 3 objectives:

1. Eliminate redundant data
2. No duplication of data means minimizing errors and anomalies
3. Keep queries execution simple for meaningful analysis

This means that one table should be dedicated to one purpose only and intermingling of various purposes in one table should be avoided. This means that each table should serve distinct and specific purpose.

If any table has the duplicate information, it is better to strip it and put it in a separate table such as product attributes in the product look up table and not combined with the order table.

**Data Tables vs. Look up Tables** 

1. **Data Tables:** Data tables are focused on providing numerical values at the granular level with key columns which can help create the relationship between tables
2. **Lookup Tables:** Lookup tables are focused on providing descriptive attributes about each dimension in a table.


**Creating Table Relationships:** There are 2 ways I can create relationships among tables in power BI:

1. Using the Model interface, I can create relationships among various tables based on primary and foreign keys.
2. I can add or detect relationships using 'Manage Relationships' dialog box.

**Star Schema vs. Snowflake Schema:** Here is the main way to identify these schemas:

- **Star Schema:** There are individual lookup tables surrounding a central data table.
- **Snowflake Schema:** Models with chains of dimensions of tables are called Snowflake schemas.

Both relationships are shown in the below picture:



**Relationship Cardinality:** Cardinality here refers to uniqueness of values in a column

- **One-to-Many:** One instance of each primary key with many instances of foreign key

We cannot have many-to-many relationships as this cannot help ensure uniqueness of primary keys. We can have one-to-one  relationship but it will be very inefficient.

**Connecting Multiple Data Tables**

In general, we should never create direct relationship between 2 data tables rather we should connect via shared lookups e.g. connection between AW_Sales data table and a new table AW_Returns data table.



**Important Reminder: We should design our models with one-way filters and one-to-many cardinality.**

## Adding New Calculated Columns Using Data Analysis Expressions (DAX)

There are 2 following ways to use DAX formula language:

1. Create Calculated Columns
2. Add measures to your modeling using intuitive syntax

**DAX Calculated Columns:** Calculated columns allow us to add new formula-based columns to tables. Calculated columns are good for adding a static fixed value to each row in a table. Do not use calculated columns for aggregation (SUM, COUNT, etc) formulas.

**Important Reminder: Calculated columns are typically used for filtering data, rather than creating numerical values.**

**DAX Measures:** Measures are DAX formulas used to generate new calculated values. Measures constantly recalculates whenever the filter-context is revised.

**Important Reminder: Use Measures to create numerical, calculated values that can be analyzed in the 'values' field of a report visual.**


**Rules to Follow for DAX vx. Measures**

- Don't use a calculated column when a measure can do the samething.
- Use calculated columns for static, fixed values
- Use Measures when aggregation is necessary and use it for dynamic values in a report
- Since a measure can be used anywhere in the report, it is a preferred solution.
- Keep measures simple so that it can be implemented easily. For complicated measures, it is a good practice to break it down to simpler components.
- Reference columns with the table name and measures alone. This helps in differentiating and it is more readable.

**Best Practices for Speed and Performance**

- Eliminate redundant columns and keep the data tables narrow
- Imported columns are better than calculated columns. If calculated columns are required, it is better to do it at the source table itself within the query editor.
- Minimize iterator functions such as Filter, SUMX etc as these are computationally intensive.
